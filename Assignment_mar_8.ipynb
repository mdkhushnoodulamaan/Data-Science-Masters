### Ans4)

The binomial distribution is a probability distribution that describes the number of successes in a fixed number of independent trials, each with the same probability of success. Here are some of the key properties of the binomial distribution:

1) The trials are independent.
2) Each trial can only result in two possible outcomes, which we can call success or failure.
3) The probability of success is the same for each trial.
4) The number of trials is fixed.
5) The random variable of interest is the number of successes in the fixed number of trials.

Two examples of events where the binomial distribution can be applied are:

    A coin-flipping experiment: Suppose we flip a fair coin 10 times and want to know the probability of getting exactly 5 heads. Each flip is independent, and there are only two possible outcomes (heads or tails) with equal probability. Thus, we can model the number of heads obtained in 10 flips using a binomial distribution.

    Quality control in manufacturing: A factory produces light bulbs, and 10% of the bulbs are defective. If a customer purchases a pack of 20 bulbs, what is the probability that at most 3 of them are defective? Each bulb in the pack is independent, and there are two possible outcomes (defective or non-defective) with different probabilities. Thus, we can model the number of defective bulbs in a pack using a binomial distribution.
