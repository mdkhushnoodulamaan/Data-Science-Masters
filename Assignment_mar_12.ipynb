```python
import scipy.stats as stats
# Set the parameters
mean = 50
std_dev = 5
n = 100 # sample size
# Calculate the standard error
std_err = std_dev / (n ** 0.5)
# Calculate the confidence interval
t_value= stats.t.ppf(0.975, n-1)
lower_ci = mean - t_value * std_err
upper_ci = mean + t_value * std_err
#print the result
print(f"The 95% confidence interval is: [{lower_ci:.2f},{upper_ci:.2f}]")
```

### Ans4)

```python
import statsmodels.api as sm

# set up the data
n = 500
smokers = 60

# calculate the point estimate and standard error of the proportion
prop = smokers / n
se = np.sqrt(prop * (1 - prop) / n)

# calculate the confidence interval
alpha = 0.05
ci = sm.stats.proportion_confint(smokers, n, alpha=alpha)

# print the results
print(f"Point estimate: {prop:.2f}")
print(f"Standard error: {se:.4f}")
print(f"{1 - alpha:.0%} Confidence interval: {ci}")
```

```python
import scipy.stats as stats

sample_mean = 75
sample_std = 12
sample_size = 50
confidence_level = 0.90

t_value = stats.t.ppf((1 + confidence_level) / 2, df=sample_size-1)
margin_of_error = t_value * (sample_std / (sample_size ** 0.5))

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("90% Confidence Interval: ({:.2f}, {:.2f})".format(lower_bound, upper_bound))
```

### Ans7)

To calculate the confidence interval for the true proportion of people in the population who prefer Coke, we can use the following formula:

CI = p ± z*(sqrt(p*(1-p)/n))

where:

p is the proportion of people in the sample who prefer Coke

z is the z-score for the desired level of confidence (in this case, 2.576 for 99% confidence)

n is the sample size

We can plug in the values we have:

p = 520/1000 = 0.52

z = 2.576

n = 1000

CI = 0.52 ± 2.576*(sqrt(0.52*(1-0.52)/1000))

Simplifying the expression under the square root:

CI = 0.52 ± 0.049

Therefore, the 99% confidence interval for the true proportion of people in the population who prefer Coke is:

CI = (0.471, 0.569)

This means that we are 99% confident that the true proportion of people in the population who prefer Coke is between 0.471 and 0.569.

### Ans10)

```python
import numpy as np
import scipy.stats as stats

# observed frequencies
observed_freq = np.array([[200, 150, 150], [225, 175, 100]])

# calculate expected frequencies assuming independence
total = np.sum(observed_freq)
row_totals = np.sum(observed_freq, axis=1).reshape(2, 1)
col_totals = np.sum(observed_freq, axis=0).reshape(1, 3)
expected_freq = row_totals * col_totals / total

# calculate chi-square test statistic
chi2_stat = np.sum((observed_freq - expected_freq)**2 / expected_freq)

# calculate degrees of freedom
dof = (2 - 1) * (3 - 1)

# calculate critical value of chi-square for alpha = 0.01
alpha = 0.01
critical_val = stats.chi2.ppf(q=1-alpha, df=dof)

# print results
print(f"chi-square statistic = {chi2_stat:.2f}")
print(f"critical value = {critical_val:.2f}")
if chi2_stat > critical_val:
    print("There is a significant association between chocolate preference and country of origin.")
else:
    print("There is no significant association between chocolate preference and country of origin.")
```

t = (sample_mean - hypothesized_mean) / (sample_standard_deviation / sqrt(sample_size))
