### Ans13)

```python
df=df[df['Type']!='0']
```

```python
df.Type.value_counts()
```

```python
chars_to_clean = ['+',',','$']
cols_to_clean = ['Installs','Price']
for item in chars_to_clean:
    for cols in cols_to_clean:
        df[cols]=df[cols].str.replace(item,'')
```

```python
df['Price']=df['Price'].astype(float)
```

```python
df.info()
```

```python
avg_price=df.groupby('Type')['Price'].mean().reset_index()
```

```python
sns.barplot(x='Price',y='Type', data= avg_price, color ='r')
```

```python
avg_price
```

- We can say the free apps have no price and paid apps have 13.92$ avg price.
